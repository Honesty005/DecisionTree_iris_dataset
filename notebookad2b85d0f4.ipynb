The goal is to create a decision tree to identify what type of flower based on the sepal's length, sepal's width and petal's length , petal's width. 

```python
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the iris dataset into pandas DataFrame
from sklearn.datasets import load_iris
iris_dataset = load_iris()
df = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)

'''
columns=iris_dataset.feature_names
Instead of just having "Column 0" or "Column 1," it uses the specific names provided by the dataset,
such as 'sepal length (cm)' and 'sepal width (cm)
'''

df.head()
df['target'] = iris_dataset.target
df
```

```python
# Extract the dependent and independent variables
X = df.drop('target', axis=1)
y = df['target']

# we define x as the depedent variables by dropping the target column
# we define y as the indepedent variables 
```

```python
# Create a decision tree classifier
dtc = tree.DecisionTreeClassifier(random_state=123)

# random_state is a seed for the pseudo-random number generator
# The number itself has no special mathematical meaning. You could use 42, 0, or 2025 and 
# the model would work exactly the same way
```

```python
# Fit the decision tree
dtc.fit(X, y)
```

```python
# from sklearn import plot_tree
# control the plotting size 
plt.figure(figsize=(10, 8))
# Plot the decision tree
tree.plot_tree(
    dtc, filled=True, feature_names=X.columns, class_names=iris_dataset.target_names
)#,max_depth=2
```

Gini Impurity measures the likelihood of a random sample being incorrectly labeled if it were classified according to the distribution of labels in a dataset.

example petal length, the Gini impurity = 1 - [(50/150)^2 + (50/150)^2 + 50/150)^2] = 0.667
to calculate the left node Gini = 1 - (50/50)^2 = 0
to calculate the right node Gini = 1 - [(0/100)^2 + (50/100)^2 + 50/100)^2] = 0.5

the higher the Gini impurity, the likelihood of the sample being incorrectly labeled.
